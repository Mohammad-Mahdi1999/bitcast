```python
import pandas as pd
```

```python
df = pd.read_csv("btcusd_1-min_data.csv")
df

```

```python
df['Datetime'] = pd.to_datetime(df['Timestamp'], unit='s')
df.set_index('Datetime', inplace=True)

# Select only the datetime and Close columns
df = df[[ 'Close']]

print(df)
```

```python
# Resample to hourly intervals and take the last value in each hour
hourly_df = df.resample('H').last()

print(hourly_df)
```

```python
import pandas as pd


# Drop rows with NaN values
Bitcoin_df = hourly_df.dropna()

print("Bitcoin Data:")
print(Bitcoin_df)


Bitcoin_df.to_csv("Bitcoin.csv")
```
